```python
import pandas as pd
import matplotlib.pyplot as plt

CSV_FILE = "r1085734.csv"
```

```python
# Salta le righe di commento e leggi solo le prime 5 colonne
df = pd.read_csv(CSV_FILE, comment='#', header=None, usecols=[0,1,2,3,4])
df.columns = ['Iteration', 'Elapsed time', 'Mean value', 'Best value', 'Cycle']

plt.figure(figsize=(10, 6))
plt.plot(df['Iteration'], df['Mean value'], label='Mean Fitness', color='blue')
plt.plot(df['Iteration'], df['Best value'], label='Best Fitness', color='red')
plt.xlabel('Iteration')
plt.ylabel('Fitness')
plt.title('Mean and Best Fitness over Iterations')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
